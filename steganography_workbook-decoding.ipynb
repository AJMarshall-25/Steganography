```python
from PIL import Image
import os
import numpy as np
```

```python
to_decode = Image.open('encoded_test.png')
```

```python
to_decode
```

```python
#convert image into a list of rgb values pixel by pixel and then converts that into a list of the valuues
image_pixels = list(to_decode.getdata())
image_lists = [list(x) for x in image_pixels]
image_list = [pix for sublist in image_lists for pix in sublist]
```

```python
# converting image_list into binary based on odd/even of rgb value

binary_values = ''
for pix in image_list:
    if pix%2 == 0:
        binary_values += '0'
    else: 
        binary_values += '1'

binary_values[:7]
```

```python
# extracting message length

mess_len = int(binary_values[0:7], 2)

mess_len
```

```python
# creating list of ascii values, since the first 7 characters are the message length we start the while loop at 7

x = 7
ascii_vals =[]
while x <= (mess_len*7):
    char = (int(binary_values[x:(x+7)], 2)+1)
    ascii_vals.append(char)
    x+= 7
    
ascii_vals
```

```python
#convert from ascii to text

message = ''
for num in ascii_vals:
    message += chr(num)
    
message
```
